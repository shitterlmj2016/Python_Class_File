```python
import requests
import bs4
```

```python
response = requests.get('https://css-tricks.com/how-css-selectors-work/')
```

```python
soup = bs4.BeautifulSoup(response.content, 'html.parser')
```

```python
all_divs = soup.select('div')
```

```python
for d in all_divs:
    print(d['class'])
```

```python
code_divs = soup.select('div.code-group')
print(code_divs)
```

```python
anchors = soup.find_all('a', href=True)
```

```python
for a in anchors:
    print(a['href'])
```
