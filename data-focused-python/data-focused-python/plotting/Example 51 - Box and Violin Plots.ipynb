```python
# standard
import numpy as np
import pandas as pd
from numpy.random import randn

# stats
from scipy import stats

# plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1, data2])
```

```python
# have whiskers include outliers
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.boxplot(data=[data1, data2], whis=np.inf, vert=False)
```

```python
# Normal Dist
data3 = stats.norm(0,5).rvs(100)

# Two gamma dist. Concatenate together
data4 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1 * stats.gamma(5).rvs(50)])

# Box plot data1 and data2
sns.boxplot(data=[data1, data2],whis=np.inf)
```
