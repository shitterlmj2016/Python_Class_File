```python
# standard
import numpy as np
import pandas as pd
from numpy.random import randn

# stats
from scipy import stats

# plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
plt.hist(dataset1, bins=100)
```

```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred')
```

```python
# normalize the datapoints because they have different numbers
plt.hist(dataset1,normed=True,color='indianred',alpha=.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
```
